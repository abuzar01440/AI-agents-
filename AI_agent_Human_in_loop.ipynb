### 🧩 **Core Components**

```python
class AgentState(TypedDict):
```

| Node | Function | Emoji | Purpose |

| `ai_agent_task` | 🤖 | ⚡ | Generates professional emails using AI |
| `human_node` | 👤 | 🧠 | Captures human feedback and decisions |
| `end_node` | 🏁 | ✨ | Finalizes and displays completed email |

````python

import datetime
from langchain.tools import tool # Import the tool decorator
from langchain_core.messages import BaseMessage, HumanMessage
from langgraph.graph import StateGraph, START, END, add_messages
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage, ToolMessage # To structure messages
from typing import List, Dict, Any, Optional, Annotated, TypedDict
from langgraph.graph import StateGraph, START, END, add_messages
from langgraph.types import Command, interrupt
from langgraph.checkpoint.memory import MemorySaver
from langchain_google_genai import ChatGoogleGenerativeAI
import uuid
print('--------------------Import Done========================')
````

````python
import getpass
import os

if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter your Google AI API key: ")
````

````python
llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-flash",
    temperature=0.7,
    max_tokens=512,
    timeout=60,
    max_retries=3
)
````

````python
class AgentState(TypedDict):
  Email_title: str
  research_interest: list[str]
  generated_first_email: Annotated[list[str], add_messages] ##Merges two lists of messages, updating existing messages by ID.
  #By default, this ensures the state is "append-only", unless the
  #new message has the same ID as an existing message.
  human_feedback: Annotated[list[str], add_messages]
  research_collected: bool
````

````python
def research_input_node(state: AgentState):
    if state.get("research_collected", False):
        return state  # Don't ask again
    research_interest = input("Please enter the research interest: ")
    return {
        "research_interest": [research_interest],
        "research_collected": True
    }
````

````python
def ai_agent_task(state: AgentState):
    print("Generating Email!")
    Email_title = state["Email_title"]
    research_interest = state["research_interest"]
    human_feedback = state.get("human_feedback", ["No feedback yet!"])

    prompt = f"""
    Write a professional email for PhD application.

    Student: Zar
    CGPA: 3.79/4
    Topic: {Email_title}
    Research: {research_interest[-1] if research_interest else "Not specified"}
    Feedback: {human_feedback[-1] if human_feedback else "No feedback"}

    Write a complete email to a professor requesting PhD acceptance letter for computer science and machine learning At BNU University, china.
    Research interests should align with professor's work.
    """

    print(f"DEBUG: Sending prompt: {prompt[:200]}...")  # Debug print

    try:
        response = llm.invoke([
            SystemMessage(content="You are an expert email writer. Write complete emails."),
            HumanMessage(content=prompt)
        ])

        print(f"DEBUG: Raw response: {response}")  # Debug print
        print(f"DEBUG: Response content: '{response.content}'")  # Debug print
        print(f"DEBUG: Response type: {type(response.content)}")  # Debug print

        generated_Email = response.content

        if not generated_Email or generated_Email.strip() == "":
            print("ERROR: No email content generated!")
            print(f"DEBUG: Response object: {response.__dict__}")  # More debug info
            generated_Email = "Error: LLM returned empty response"
        else:
            print(f"[ai_agent_task] Generated email:\n{generated_Email}\n")

    except Exception as e:
        print(f"ERROR: Exception during LLM call: {e}")
        generated_Email = f"Error: {str(e)}"

    return {
        "generated_first_email": state["generated_first_email"] + [generated_Email]
    }
````

````python
# interrupt, Human LOop
def human_node(state: AgentState):
  """here, we will add human interference for desire email generator"""
  print("Human Interference!")

  generated_first_email = state["generated_first_email"]

  user_feedback = interrupt({
      "generated_first_email" : generated_first_email,
      "message" : "Provide feedback or type 'done' or go ahead to finish"})
  print(f"[human_node] Received human feedback: {user_feedback}")

  # if user write done or go head, then it will to the end node
  if user_feedback.lower() in ["done", "go ahead"]:
        return Command(update={"human_feedback": state["human_feedback"] + ["Finalised"]}, goto="end_node")

    # Otherwise, update feedback and return to model for re-generation
  return Command(update={"human_feedback": state["human_feedback"] + [user_feedback]}, goto="ai_agent_task")
````

````python
def end_node(state: AgentState):
    print("\n" + "="*60)
    print("FINAL EMAIL:")
    print("="*60)
    if state["generated_first_email"]:
        print(state["generated_first_email"][-1])
    return state
````

````python
graph = StateGraph(AgentState)
graph.add_node(research_input_node, name="research_input_node")
graph.add_node(ai_agent_task, name="ai_agent_task")
graph.add_node(human_node, name="human_node")
graph.add_node(end_node, name="end_node")

graph.set_entry_point("research_input_node")

graph.add_edge("research_input_node", "ai_agent_task")
graph.add_edge("ai_agent_task", "human_node")
graph.add_edge("human_node", "ai_agent_task")

graph.set_finish_point("end_node")


# Enable Interrupt mechanism
checkpointer = MemorySaver()
app = graph.compile(checkpointer=checkpointer)

````

````python
thread_config = {"configurable": {"thread_id": str(uuid.uuid4())}}

while True:  # Main loop - continues until user types 'exit'
    print("\n" + "="*60)
    print("NEW EMAIL GENERATION SESSION")
    print("="*60)

    email_title = input("Enter your Email topic (or 'exit' to quit): ")

    # Check if user wants to exit
    if email_title.lower() == 'exit':
        print("Goodbye!")
        break

    # Create new initial state for each session
    initial_state = {
        "Email_title": email_title,
        "generated_first_email": [],
        "human_feedback": [],
        "research_collected": False
    }

    # Generate new thread ID for each session
    thread_config = {"configurable": {"thread_id": str(uuid.uuid4())}}

    print(f"\nStarting email generation for: {email_title}")

    # Inner loop for feedback iterations
    session_complete = False
    while not session_complete:
        try:
            for chunk in app.stream(initial_state, config=thread_config):
                for node_id, value in chunk.items():
                    if node_id == "__interrupt__":
                        print("\n" + "-"*50)
                        print("FEEDBACK OPTIONS:")
                        print("- Provide specific feedback to improve the email")
                        print("- Type 'done' or 'go ahead' to finish this email")
                        print("- Type 'restart' to start over with new topic")
                        print("- Type 'exit' to quit completely")
                        print("-"*50)

                        user_feedback = input("Your input: ")

                        # Handle different user inputs
                        if user_feedback.lower() == 'exit':
                            print("Goodbye!")
                            session_complete = True
                            exit()  # Exit completely

                        elif user_feedback.lower() == 'restart':
                            print("Restarting with new email topic...")
                            session_complete = True
                            break  # Break to outer while loop for new email

                        elif user_feedback.lower() in ["done", "go ahead"]:
                            print("Email finalized!")
                            app.invoke({"human_feedback": [user_feedback]}, config=thread_config)
                            session_complete = True
                            break

                        else:
                            print(f"Processing feedback: {user_feedback}")
                            # Continue with feedback - update the initial_state for next iteration
                            app.invoke({"human_feedback": [user_feedback]}, config=thread_config)
                            # Don't break - continue the feedback loop

        except KeyboardInterrupt:
            print("\n\nSession interrupted. Starting new session...")
            break
        except Exception as e:
            print(f"Error occurred: {e}")
            print("Starting new session...")
            break

    if session_complete:
        print("\n" + "="*60)
        print("EMAIL SESSION COMPLETED")
        print("="*60)
        continue_session = input("Do you want to create another email? (y/n): ")
        if continue_session.lower() in ['n', 'no']:
            print("Goodbye!")
            break
````
